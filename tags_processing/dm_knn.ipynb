```python
# knn on datasets
import pandas as pd
df = pd.read_csv("dm1/dm_clean2.csv")
```

```python
df_clusters = pd.read_csv("dm1/all_clusters.csv")
```

```python
df_clusters
```

```python
from sklearn.neighbors import NearestNeighbors 
import numpy as np
X = df_clusters.as_matrix()
nbrs = NearestNeighbors(n_neighbors=1,algorithm='ball_tree').fit(X)
```

```python
distances , indices = nbrs.kneighbors(X)
```
