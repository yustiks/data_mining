```python
# load latitude and longitude
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from collections import Counter
import pandas as pd, numpy as np, matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
#from geopy.distance import great_circle
#from shapely.geometry import MultiPoint
df = pd.read_csv('dm1/dm_clean1.csv')
```

```python
X=df.loc[:,['latitude','longitude']]
```

```python
# 
id_n=10
kmeans = KMeans(n_clusters=id_n, random_state=42).fit(X)
id_label=kmeans.labels_
```

```python
print(id_label)
```

```python
#plot result
ptsymb = np.array(['b.','r.','m.','g.','c.','k.','b*','r*','m*','r^']);
plt.figure(figsize=(12,12))
plt.ylabel('Longitude', fontsize=12)
plt.xlabel('Latitude', fontsize=12)
for i in range(id_n):
    cluster=np.where(id_label==i)[0]
    plt.plot(X.longitude[cluster].values,X.latitude[cluster].values,ptsymb[i])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.vq import kmeans2, whiten

x, y = kmeans2(X, 500, iter = 20)  #<--- I randomly picked 500 clusters
plt.scatter( X.loc[:,'longitude'], X.loc[:,'latitude'], c=y, alpha=0.33333);

```
